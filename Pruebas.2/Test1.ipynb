Modulos usados

```python
from ecg_class import ECGClass
from series_v2 import timeSerie
import numpy as np
```

Obtengo datos del ECG para poder verificar el funcionamiento de timeSerie, para eso cargo un array con una columna de tiempos y una columna de valores de parametros, ya sea intervalos RR, derivada máxima o angulo formado entre la máxima y la minima derivada del complejo QRS 

```python
ECG = ECGClass('105')
RR  = ECG.intervalRR()
ANG = ECG.angle()
SM  = ECG.slopeMax()
matrixECG = np.column_stack((ECG.time, ECG.signal))
```

Genero una instancia TimeSerie con los valores de los intervalos RR. Si se selecciona cualquiera de los punto se genera una nueva gráfica mostrando el complejo QRS que dio origen a esta muestra

```python
pruebaRR = timeSerie('IntervalRR', matrixECG, RR, ECG.qrs/ECG.fs)
pruebaRR.plotSerie()
```

Genero una instancia TimeSerie con los valores de las pendientes máximas. Si se selecciona cualquiera de los punto se genera una nueva gráfica mostrando el complejo QRS que dio origen a esta muestra

```python
pruebaSM = timeSerie('SlopeMax', matrixECG, SM[:,1], SM[:,0]  )
pruebaSM.plotSerie()
```

Utilizo una funcion que permite recortar los datos si estos exeden un intervalo determinado, generando una nueva instancia TimeSerie, con la cual se va a poder inspeccionar la TimeSerie que le dio origen y desde ella el ECG.
En este caso toma como origen todo el TimeSerie anterior, pero en otro tipo de funciones puede ser conveniente que el origen de esta nueva instancia esté limitado a solo un rango de valores de la instancia "padre"

```python
rectYpruebSM = pruebaSM.recorteY(25,30)
rectYpruebSM.plotSerie()
```

Por último pruebo los datos de los angulos y además de utilizar la funcion de acortar la serie segun los valores, también la acoto en tiempo, generando dos instancias distintas por las cuales se puede inspeccionar

```python
pruebaANG = timeSerie('Angle', matrixECG, ANG[:,1], ANG[:,0],  ANG[:,3], ANG[:,2], ANG[:,5], ANG[:,4] )
pruebaANG.plotSerie()

rectXpruebANG = pruebaANG.recorteX(0,20)
rectXpruebANG.plotSerie()

rectYpruebANG = rectXpruebANG.recorteY(4,5)
rectYpruebANG.plotSerie()
```
